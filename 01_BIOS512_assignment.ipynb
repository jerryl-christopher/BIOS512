Using the `penguins` dataset, make a scatter chart with `bill_length_mm` on the x-axis and `bill_depth_mm` on the y-axis.

```r
penguins |> ggplot(
    mapping = aes(x = bill_length_mm, y = bill_depth_mm)) +
    geom_point()+
    labs(
    title = "Bill Length and Bill Depth of Penguins",
    subtitle = "In 3 Species of Penguins",
    x = "Bill Length (mm)",
    y = "Bill Depth (mm)"
  )
```

Make the same plot but *encode* the `species` column with color.

```r
penguins |> ggplot(
    mapping = aes(x = bill_length_mm, y = bill_depth_mm)) +
    geom_point(mapping = aes(color = species))+
    scale_color_brewer(palette = 'Set2')+
    labs(
    title = "Bill Length and Bill Depth of Penguins",
    subtitle = "In 3 Species of Penguins",
    x = "Bill Length (mm)",
    y = "Bill Depth (mm)",
    color = "Species"
  )
```
